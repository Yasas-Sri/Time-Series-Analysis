```python
#import relevant libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the data
url = 'https://raw.githubusercontent.com/Yasas-Sri/Time-Series-Analysis/refs/heads/main/train.csv'
df = pd.read_csv(url)
```

# Exploratory data analysis



```python


# Examine Total Sales by category
sales_by_category = df.groupby('Category', observed=False)['Sales'].sum()
plt.figure(figsize=(12, 4))
plt.bar(sales_by_category.index, sales_by_category.values)
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python


# Examine Total Sales by sub-category
sales_by_subcategory = df.groupby('Sub-Category', observed=False)['Sales'].sum()
plt.figure(figsize=(12, 4))
plt.bar(sales_by_subcategory.index, sales_by_subcategory.values)
plt.title('Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```
